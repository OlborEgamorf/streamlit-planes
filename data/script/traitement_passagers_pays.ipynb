```python
import pandas as pd
```

```python
file = "../passagers_par_pays_traite.csv"

df = pd.read_csv(file, delimiter=",")

print(df.head())

print(len(df))
```

```python
file2 = "../vols_par_pays_traite.csv"

df_vols = pd.read_csv(file2, delimiter=",")

df_selected = df_vols[['COUNTRY_ID', 'COUNTRY_NAME']]

print(df_selected.head())
```

```python
# Convertir le DataFrame en dictionnaire
dict_selected = df_selected.to_dict(orient='records')

# Afficher le dictionnaire
dict_selected = {
    item['COUNTRY_ID']: item['COUNTRY_NAME'] for item in dict_selected
}
```

```python
df["COUNTRY_NAME"] = df['COUNTRY_ID'].map(dict_selected)
print(df.head(20))
```

```python
df.to_csv("../passagers_par_pays_traite2.csv", index=False)
```
