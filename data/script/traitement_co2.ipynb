```python
import pandas as pd
from list_iso import *
```

```python


file = "../annual-co-emissions-from-aviation.csv"

df = pd.read_csv(file, delimiter=",")

print(df.head())

print(len(df))
```

```python
file2 = "../vols_par_pays_traite.csv"

df_vols = pd.read_csv(file2, delimiter=",")

df_selected = df_vols[['COUNTRY_ID']].drop_duplicates()

print(df_selected.head(20))

df_selected.to_dict(orient="list")


```

```python
#Traitement des données 
# Suppression des continents non nécessaires
df = df[~df['Entity'].isin(['World', 'Africa', 'Asia', 'Europe', 'European Union (27)', 'North America', 'Oceania', 'South America'])]

print(len(df))
```

```python
df.sort_values(by=['Entity', 'Year'], inplace=True)
print(df.head(20))
```

```python
list_iso_key = convert_ISO_3166_2_to_1.keys()
list_iso_values = convert_ISO_3166_2_to_1.values()

print(list_iso_key)
print(list_iso_values)
```

```python
inverted_dict = {v: k for k, v in convert_ISO_3166_2_to_1.items()}
print(inverted_dict)
```

```python
df["COUNTRY_ID"] = df['Code'].map(inverted_dict)
print(df.head())
```

```python
df = df[df["COUNTRY_ID"].isin(df_selected["COUNTRY_ID"])]
```

```python
df.rename(columns={"Entity": "COUNTRY_NAME", "Year": "YEAR", "Total annual CO₂ emissions from aviation": "CO2_EMISSIONS_TONNES"}, inplace=True)

# Sauvegarde
df.to_csv("../co2_par_pays.csv", index=False)
```
