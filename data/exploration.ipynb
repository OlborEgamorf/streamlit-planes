```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Import des données

```python
data_vols_pays = pd.read_csv("vols_par_pays_traite.csv")
data_vols_aeroport = pd.read_csv("vols_par_aeroport_traite.csv")
data_passagers_aeroport = pd.read_csv("passagers_par_aeroport_traite.csv")
data_passagers_pays = pd.read_csv("passagers_par_pays_traite2.csv")
data_co2_pays = pd.read_csv("co2_par_pays.csv")
```

Graphiques pour analyser les données

```python
plt.figure(figsize=(12,6))
sns.barplot(data=data_vols_pays, x='COUNTRY_NAME', y='ARRIVAL_VALUE')
plt.xticks(rotation=90)
plt.title("Nombre de vols (arrivés) par pays")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(data=data_passagers_pays, x='COUNTRY_NAME', y='DEPARTURE_VALUE')
plt.xticks(rotation=90)
plt.title("Nombre de passagers (départs) par pays")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(data=data_co2_pays, x='YEAR', y='CO2_EMISSIONS_TONNES', hue='COUNTRY_NAME')
plt.title("Relation entre année et émissions de CO2")


plt.legend(title='Pays', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout() 
plt.show()
```

```python
# Choisir le pays à afficher
pays_selection = "FR"

# Filtrer les données
df_filtre = data_passagers_aeroport[data_passagers_aeroport["COUNTRY_ID"] == pays_selection]

# Affichage du graphique
plt.figure(figsize=(12,6))
sns.barplot(data=df_filtre, x='AIRPORT_NAME', y='DEPARTURE_VALUE')
plt.xticks(rotation=90)
plt.title(f"Nombre de passagers par aéroport en {pays_selection}")
plt.show()
```
